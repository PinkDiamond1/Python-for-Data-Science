```python
# Import the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the mpg dataset that comes with seaborn. This dataset contains technical specifications of cars
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
# Some more info on the dataset: check for missing values, data types, number of observations
mpg.info()
```

```python
# Distribution of values across different characteristics 
mpg.describe()
```

```python
## https://seaborn.pydata.org/examples/different_scatter_variables.html
    
sns.set(style="whitegrid")

f, ax = plt.subplots(figsize=(14, 6.5))

sns.scatterplot(x="mpg", 
                y="cylinders",
                palette="Blues",
                linewidth=0,
                data=mpg, 
                ax=ax)
```

```python
## https://seaborn.pydata.org/examples/grouped_boxplot.html

f, ax = plt.subplots(figsize=(14, 8))

sns.boxplot(x = "cylinders", 
            y = "mpg", 
            data = mpg, 
            whis = 100, 
            palette = 'Reds')
```

```python
## https://seaborn.pydata.org/generated/seaborn.violinplot.html

f, ax = plt.subplots(figsize=(14, 6))

sns.violinplot(x="cylinders", 
               y="mpg", 
               data = mpg, 
               palette = "Blues", 
               inner = "point") # or, None, or "box", or "quartile"

sns.despine(left=False)
```

### Two series comparison using hue

```python
## https://seaborn.pydata.org/generated/seaborn.violinplot.html
f, ax = plt.subplots(figsize=(14, 6))

sns.violinplot(x="cylinders", 
               y="mpg", 
               hue = 'origin',
               data = mpg[(mpg.origin=="europe") | (mpg.origin=="japan")], 
               palette = "Reds",
               scale = 'count',
               split = True,
               inner = None) # or, None, or "box", or "quartile"
```

```python
## https://seaborn.pydata.org/examples/different_scatter_variables.html
    
sns.set(style="whitegrid")

f, (ax1, ax2) = plt.subplots(1,2,figsize=(14, 6.5),sharey=True)

sns.scatterplot(x="mpg", 
                y="horsepower",
                palette="Blues",
                linewidth=0,
                data=mpg, 
                ax=ax1)

sns.scatterplot(x="weight", 
                y="horsepower",
                palette="Blues",
                linewidth=0,
                data=mpg, 
                ax=ax2)
```

```python
## https://seaborn.pydata.org/generated/seaborn.relplot.html


g= sns.relplot(x="horsepower", 
            y="mpg", 
            hue="origin", 
            size="weight",
            sizes=(40, 200), 
            alpha=0.8, 
            palette="muted",
            height = 9, 
            data = mpg)

g.fig.suptitle("Horsepower vs. mpg")
```
