# Lab 1: Introducing Code Notebooks

Notebooks are documents comprising code together with other elements: text, equations and charts.

Why are they so popular in data science? They provide an easy-to-use coding environment. They also make it easy to share work with colleagues. Your annotated notebook is easy for colleagues to understand, because intermediate outputs (eg. charts) are displayed alongside the code that generated them.

#### The Notebook environment
To edit notebooks, use Jupyter Notebooks (part of the Anaconda distribution of Python) or Colab (a free online platform from Google). They both work in your web browser. Notebooks can be saved as .ipynb files. They comprise two types of cells:
* _Code cells._ These contain executable commands in Python.
* _Text cells._ These include plain text, or you can use [markdown](https://commonmark.org/help/) to add formatting.
